```python
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "ASSIGNMENT 4/schools_complete.csv"
student_data_to_load = "ASSIGNMENT 4/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
#Calculate the total number of schools
skl_tot = len(school_data_complete["school_name"].unique())

#Calculate the total number of students
student_tot = school_data_complete["Student ID"].count()
```

```python
#Calculate the total budget
budget_tot = school_data_complete["budget"].sum()
```

```python
#Calculate the average math score 
math_avg = school_data_complete["math_score"].mean()

#Calculate the average reading score
reading_avg = school_data_complete["reading_score"].mean()
```

```python
# Calculate the percentage of students with a math score of (70 or greater)
students_passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
number_students_passing_math = students_passing_math["Student ID"].count()
MathPass = (number_students_passing_math / student_tot) * 100


# Calculate the percentage of students with a reading score of (70 or greater)
students_passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
number_students_passing_reading = students_passing_reading["Student ID"].count()
readingPass = (number_students_passing_math / student_tot) * 100


# Calculate the percentage of students who passed math and reading (% Overall Passing)
overallPass = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['Student ID'].count() / student_tot *100

```

```python
#Displaying the output in a summary table
# Create a dataframe to hold the above results
District_summary = pd.DataFrame({
    "Total Schools": skl_tot,
    "Total Students": f"{student_tot:,}",
    "Total Budget": f"£{budget_tot:,.2f}",
    "Average Math Score": f"{math_avg:.6f}",
    "Average Reading Score": f"{reading_avg:.5f}",
    "% Passing Math": f"{MathPass:.6f}",
    "% Passing Reading": f"{readingPass:.6f}",
    "% Overall Passing Rate": f"{overallPass:.6f}"
}, index=[0])


#Display
District_summary.head()
```

```python
# Group by school name
school_name = school_data_complete.set_index("school_name").groupby(["school_name"])

# school types by school name
skl_type = school_data.set_index("school_name")["type"]
```

```python
# Calculate total students
student_tot = school_name['Student ID'].count()
```

```python
# Total school budget
skl_budget = school_data.set_index("school_name")["budget"]

# Budget per student
budget_per_student = (school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"])
```

```python
# Average Math Score
average_math = school_name['math_score'].mean()

# Average Reading Score
average_reading = school_name['reading_score'].mean()
```

```python
# % Passing Math
math_percent = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/student_tot*100

# % Passing Reading
read_percent = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/student_tot*100

# % Overall Pass
overall_percent = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/student_tot*100
```

```python
skl_summary = pd.DataFrame({
    "School Type": skl_type,
    "Total Students": student_tot,
    "Per Student Budget": budget_per_student,
    "Total School Budget": skl_budget,
    "Average Math Score": average_math,
    "Average Reading Score": average_reading,
    '% Passing Math': math_percent,
    '% Passing Reading': read_percent,
    "% Overall Passing": overall_percent})
```

```python
#formatting

#munging
skl_summary = skl_summary[['School Type', 
                          'Total Students', 
                          'Total School Budget', 
                          'Per Student Budget', 
                          'Average Math Score', 
                          'Average Reading Score',
                          '% Passing Math',
                          '% Passing Reading',
                          '% Overall Passing']]



skl_summary.style.format({'Total Students': '{:}',
                          "Total School Budget": "£{:,.2f}",
                          "Per Student Budget": "£{:.2f}",
                          'Average Math Score': "{:6f}", 
                          'Average Reading Score': "{:6f}", 
                          "% Passing Math": "{:6f}", 
                          "% Passing Reading": "{:6f}"})
```

```python
# create spending bins
bins = [0, 584, 629, 644, 675]
group_name = ["=<£584", "£585-629", "£630-644", "£645-675"]

school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_name)
```

```python
#group by spending
by_spending = school_data_complete.groupby('spending_bins')
avg_math = by_spending['math_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()*100
```

```python
# Creating the dataframe           
scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "% Overall Passing": overall})
```

```python
#reorder columns
scores_by_spend = scores_by_spend[[
    "Average Math Score",
    "Average Reading Score",
    "% Passing Math",
    "% Passing Reading",
    "% Overall Passing"]]

scores_by_spend.index.name = "Per Student Budget"
```

```python
#formating
scores_by_spend.style.format({'Average Math Score': '{:.2f}', 
                              'Average Reading Score': '{:.2f}', 
                              '% Passing Math': '{:.2f}', 
                              '% Passing Reading':'{:.2f}', 
                              '% Overall Passing': '{:.2f}'})
```

```python
# create size bins
bins = [0, 1000, 1999,5000]
group_name = ["Small (<1000)", "Medium (1000-2000)" , "Large (2000-5000)"]
school_data_complete['size_bins'] = pd.cut(school_data_complete['size'], bins, labels = group_name)   
```

```python
#group by spending
by_size = school_data_complete.groupby('size_bins')
```

```python
#calculations 
average_math_score = by_size['math_score'].mean()
average_reading_score = by_size['math_score'].mean()
pass_math_percent = school_data_complete[school_data_complete['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()*100
pass_read_percent = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()*100
```

```python
# df build            
scores_by_size = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': pass_math_percent,
    '% Passing Reading': pass_read_percent,
    '% Overall Passing': overall})
```

```python
#ordering the columns
scores_by_size = scores_by_size[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    '% Overall Passing']]

scores_by_size.index.name = "Total Students"
scores_by_size = scores_by_size.reindex(group_name)

#formating
scores_by_size.style.format({'Average Math Score': '{:.6f}', 
                              'Average Reading Score': '{:.6f}', 
                              '% Passing Math': '{:.6f}', 
                              '% Passing Reading':'{:.6f}', 
                              '% Overall Passing': '{:.6f}'})
```
